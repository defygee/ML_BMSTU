#### Также студентов можно разделить на группы по их данным (алгоритм кластеризации)

```python
g = (sns.jointplot("GRE Score", "Chance of Admit ",
                   data=df, color="k")
        .plot_joint(sns.kdeplot, zorder=0, n_levels=6))
```

```python
g = sns.jointplot("TOEFL Score", "Chance of Admit ", data=df,
                  kind="kde", space=0, color="g")
```

```python
g = sns.jointplot("TOEFL Score", "GRE Score", data=df, kind="hex")
```

```python
g = sns.jointplot("GRE Score", "University Rating", data=df, kind="hex")
```
